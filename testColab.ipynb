```python
dfo.loc[(dfo["height_cm"] > 210) & (dfo["born_country"] == "USA"), :] # select rows where height_cm > 220 and born_country is ESP
```

```python
dfo[dfo["name"].str.contains("Smith")].sample(7)  # select rows where 'name' contains 'Smith'
```

```python
dfo[(dfo["born_country"].isin(["USA", "CAN", "MEX"]) ) & (dfo.height_cm > 190) & (dfo.weight_kg.notna()) ]  # select rows where 'born_country' is in the specified list
```

```python
dfo.query("height_cm > 200 and born_country == 'USA' and weight_kg.notna()")  # select rows where height_cm > 200 and born_country is USA using query method
```

```python
dfo["is_died_date_null"] = dfo["died_date"].isnull()  # create a new boolean column indicating if 'died_date' is null
```

```python
dfo.head()
```

```python
dfo_copy = dfo.copy()  # create a copy of the DataFrame to avoid SettingWithCopyWarning
# when we doon't use .copy for creating a copy of dfo both will point to the same place in the memory so if we perform an action on one the other with have the concequencies too
dfo.drop(columns=["is_died_date_null"], inplace= True)
```

```python
dfo_copy.columns
```

```python
dfo.columns
```

```python
dfo_copy.rename(columns={"is_died_date_null": "died_date_null"}, inplace=True)
```

```python
import numpy as np
dfo_copy["born_country"] = np.where(dfo_copy["born_country"] == "FRA","Fr", dfo_copy["born_country"])  # replace 'FRA' with 'Fr' in 'born_country' column 
# it is equivalent in pandas to dfo_copy["born_country"] = dfo_copy["born_country"].where(dfo_copy["born_country"] != "FRA", "Fr")

# or more clean dfo_copy["born_country"] = dfo_copy["born_country"].replace("FRA", "Fr")
```
